```python
#Feature extraction for Dog images 
i=0
for filename in os.listdir(folder1):
    path=os.path.join(folder1,filename)
    a=cv2.imread(path)
    
    #resize image
    resize=(512,512)
    img=cv2.resize(a,resize)#resize image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    #ORB feature extraction
    orb = cv2.ORB_create()
    keypoints, descriptors = orb.detectAndCompute(gray, None)
    
    #convert the descriptor array into a dataframe format
    out=pd.DataFrame(descriptors)
    print("descriptor shape ",i," : ", out.shape)
    i=i+1
    
    #append to the csv file
    csv_data=out.to_csv('ORB_dog.csv', mode='a', header=False,index=False)
```

## Training and Testing

```python
#assigning x the columns from 0 to 5 for training
x = data.iloc[:, 0:5]
print("X values")
print(x)

#assigning y with the column "Class" as target variable
y = data.iloc[:,5]
print("Y values")
print(y)
```

```python
#standard scaling

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x= scaler.fit_transform(x)
```

```python
#Dataset split into train and test with 80% Training and 20% Testing
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.20, random_state=0)
```

```python
#library
from sklearn import neighbors
from sklearn.metrics import mean_squared_error 
from math import sqrt
import matplotlib.pyplot as plt
```

```python
WCSS = [] #to store rmse values for different k
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsClassifier(n_neighbors = K)

    model.fit(x_train, y_train) 
    pred=model.predict(x_test) 
    error = sqrt(mean_squared_error(y_test,pred)) 
    WCSS.append(error)
```

```python
#Elbow curve

#elbow curve
plt.figure(figsize=(10,6))
plt.plot(range(1,21),WCSS,color='blue', linestyle='dashed', marker='o',markerfacecolor='red', markersize=10)
plt.title('WCSS Vs K')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import accuracy_score


#Create a svm Classifier
model = neighbors.KNeighborsClassifier(n_neighbors = 11)

#Train the model using the training sets
model.fit(x_train,y_train)

#Predict the response for test dataset
pred=model.predict(x_test) 
#Results
print("KNN Results")
print("KNN: Accuracy: ",accuracy_score(y_test, pred)*100,"%")
```
