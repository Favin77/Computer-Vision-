## Hough Transform

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read the image in gray format
img = cv2.imread(r'C:\Users\francis\Desktop\2.png',0)
img= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#resize image
resize=(512,512)
img=cv2.resize(img,resize)

plt.imshow(img)
plt.title("Original")
plt.show()
```

```python
#Canny edge detection to input edges for hough lines
edges = cv2.Canny(img,50,150)

plt.imshow(edges)
plt.title("Canny")
plt.show()
```

```python
# lines returns an array of rho and theta values 
lines = cv2.HoughLines(edges,1,np.pi/180, 200)

#cv2.HoughLines(edges,1,np.pi/180, 200) 
#edges are the canny edges detected
#1 = rho value(resolution in pixels)
#np.pi/180(resolution in radian) = theta value
#150 = Threshold, minimum no of intersection to detect a line

```

```python
# The below for loop runs till r and theta values, in the form of a 2-D array 

for i in range(0,len(lines)):
    #first element is an array of 2 elements rho and theta  
    r=lines[i][0][0] 
    t=lines[i][0][1] 
    
    #value of cos(t)
    a = np.cos(t)
  
    #value of sin(t)
    b = np.sin(t)
      
    #value rcos(t)
    x0 = a*r
      
    #value rsin(t)
    y0 = b*r
      
    #rcos(t)-1000sin(t), round of value it will store
    x1 = int(x0 + 1000*(-b))
      
    #rsin(t)+1000cos(t)
    y1 = int(y0 + 1000*(a))
  
    #rcos(t)+1000sin(t)
    x2 = int(x0 - 1000*(-b))
      
    #rsin(t)-1000cos(t)
    y2 = int(y0 - 1000*(a))
      
    #line drawn from (x1,y1) to (x2,y2).
    cv2.line(img,(x1,y1), (x2,y2), (0,255,0),2)
    
    
```

```python
plt.imshow(img)
plt.title("Hough Transform")
plt.show()
```
